### TASK: Recreate the Scatter Plot

**The scatterplot attempts to show the relationship between days employed and a person's age (DAYS_BIRTH) for everyone except the unemployed. Note that for this graph you will need to first remove unemployed people from the data set. Also note the values on the axes - the values have been converted to be positive. You can also change the *alpha* and *linewidth* parameters if you wish, since there are many points overlapping each other.**

```python
employees = df[df['DAYS_EMPLOYED']<0]
```

```python
employees['DAYS_EMPLOYED'] = employees['DAYS_EMPLOYED'] * -1 
```

```python
employees["DAYS_BIRTH"] = employees["DAYS_BIRTH"] * -1
```

```python
plt.figure(dpi=200)
sns.scatterplot(data = employees, x = "DAYS_BIRTH", y = 'DAYS_EMPLOYED', alpha=0.01)
```

```python
df['years'] = df['DAYS_BIRTH'].apply(lambda x : round(x/365 * -1))
```

```python
sns.histplot(data=df, x="years", color = 'red', binwidth=1)
```

### TASK: create graph Categorical:


**This graph only shows information for *half of all people* with income in this data set. Moreover, we take the “bottom” half - those who earn less. Boxplot plots are displayed for each value of the NAME_FAMILY_STATUS column, showing the distribution of their total income. Hue coloring is based on the "FLAG_OWN_REALTY" column. Note: to take only the bottom half of the income distribution, you will need to transform the dataframe or take only part of it *before* drawing the graph.**

```python
df_small = df.nsmallest(int(len(df)/2), 'AMT_INCOME_TOTAL')
```

```python
plt.figure(dpi=200, figsize=(10,5))
sns.boxplot(data = df_small, x= 'NAME_FAMILY_STATUS', 
            y = 'AMT_INCOME_TOTAL', hue = 'FLAG_OWN_REALTY')
plt.legend(loc='best', bbox_to_anchor=(1.2,0.5))
```

### TASK: Recreate the Heat Map graph:


**This heatmap shows the correlation between columns in a dataset. Correlation can be calculated using the .corr() method. Therefore, you should remove the FLAG_MOBIL column before plotting.**
