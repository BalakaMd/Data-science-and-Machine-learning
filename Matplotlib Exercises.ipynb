A "yield curve" in finance is a graph showing yields or interest rates as a function of contract length (2 months, 2 years, 20 years, etc.). The dollar interest paid on U.S. Treasury securities for different maturities is of interest to traders, and this chart is informally called "the yield curve."

## Assignment 2, part 1: Data for the yield curve

**We took for you the necessary data for the yield curve from the website [US Treasury Dept.](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx? data=yield) The data shows US bond interest rates for various contract terms. The labels list displays the contract terms for the corresponding items in the lists of values.**

**Assignment: Run the cell below to create lists for later plotting.**

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

**Challenge: Figure out how you can draw both of these lines on the same graph. Add a legend to the chart.**

```python
plt.plot(labels, july16_2007, label="july16_2007")
plt.plot(labels, july16_2020, label="july16_2020")
plt.legend()
```

**Challenge: In the graph above, the legend looks a little strange - although it doesn't block the graph, it would be better to place the legend *outside* the graph. Find a way to move the legend so that it is on the outside.**

```python
plt.plot(labels, july16_2007, label="july16_2007")
plt.plot(labels, july16_2020, label="july16_2020")
plt.legend(bbox_to_anchor=(1.3, 0.5));
```

**Extra Bonus Challenge: Try reproducing the graph below, with two axes. Although two different axes are not really needed in this example, it is a good Matplotlib exercise that will be useful in other situations.**

```python
fig, axes1 = plt.subplots()
axes1.plot(labels, july16_2007, label="july16_2007")
axes1.set_ylabel("2007")
for label in axes1.get_yticklabels():
    label.set_color("blue")

axes2=axes1.twinx()
axes2.plot(labels, july16_2020, label="july16_2020", color="red")
axes2.set_ylabel("2020")
for label in axes2.get_yticklabels():
    label.set_color("red")
```
